```python
import json
import pandas as pd
```

```python
data = pd.read_json('sampled_data.jsonl', lines=True)
data
```

```python
df = data.copy()
```

```python
s = df['datePublished']
s = s.str.split('-', expand=True)
df['year'] = s[0]
df
```

```python
# There are 1500 articles taken from a 115 year period
years = df['year'].value_counts().to_dict()
year_keys = list(years.keys())
year_keys.sort()
print(len(year_keys))
year_sort = {i: years[i] for i in year_keys}
print(year_sort)
```

```python
# This is the function that is necessary to multiply the elements into strings
text = []
def texter(column):
    uni = column
    text = []
    for i in uni:
        counter = 0 
        while counter < uni[i] :
            text.append(i)
            counter += 1
    return text
```

```python
df['unitext'] = df['unigramCount'].apply(texter)
df['bitext'] = df['bigramCount'].apply(texter)
df['tritext'] = df['trigramCount'].apply(texter)
df['text'] = df['unitext'] + df['bitext'] + df['tritext']
df['unicount'] = df['unitext'].str.len()
df['bicount'] = df['bitext'].str.len()
df['tricount'] = df['tritext'].str.len()
df['textcount'] = df['text'].str.len()
df['checksum'] = df['unicount'] + df['bicount'] + df['tricount']
df
print(df['textcount'].sum())
df
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
stop_words = set(stopwords.words('english'))
df['filtered_text'] = df['text'].apply(lambda x: [item.lower() for item in x if item.lower() not in stop_words])
second_stop = ['university', 'pp', 'historical']
df['filtered_text'] = df['text'].apply(lambda x: [item.lower() for item in x if item.lower() not in second_stop])
df
```

```python
df['filtered_string'] = [' '.join(map(str, l)) for l in df['filtered_text']]
df['filtered_string'] = df.filtered_string.str.replace('[^a-zA-Z ]', '')
df
```

```python
df['decade']=df['year']
df['decade'] = df['decade'].replace({'^190':'1900s','^191':'1910s','^192':'1920s',
                                     '^193':'1930s','^194':'1940s','^195':'1950s',
                                     '^196':'1960s','^197':'1970s','^198':'1980s',
                                    '^199':'1990s','^200':'2000s','^201':'2010s',
                                    '^202':'2020s',}, regex=True)
df['decade']=df['decade'].str[:-1]
df
```

```python
decade = '1960'
plural = decade + 's'
dy = df[df['decade']==plural]
totaltext = ', '.join(dy.filtered_string)
print(len(totaltext))
text_file = open(decade + ".txt", "w")

text_file.write(totaltext)

text_file.close()
del totaltext
```
